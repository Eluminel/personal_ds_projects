```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd

from scipy.stats import ttest_ind
from scipy.stats import f_oneway, shapiro

pokemon = pd.read_csv('https://raw.githubusercontent.com/a-milenkin/datasets_for_t-tests/main/pokemon.csv', on_bad_lines='skip')  # Откроем датасет
pokemon.head(30)
```

```python
#Task 1
#Professor Oak suspects that Grass-class Pokémon have a stronger Normal Attack than Rock-class Pokémon. Check if he is right and statistically prove your conclusion.
#Note: If there are Pokémon that are both classes, just throw them away;
#You can assume that the distribution of Normal Attacks is normal for all Pokémon classes.
```

```python
#Remove rows with Pokemon that belong to both classes
pokemon_1 = pokemon[~((pokemon['Class 1'] == 'Grass') & (pokemon['Class 2'] == 'Rock'))]
pokemon_1 = pokemon[~((pokemon['Class 1'] == 'Rock') & (pokemon['Class 2'] == 'Grass'))]
```

```python
#Select data for the Grass and Rock classes
grass_attack = pokemon_1.loc[pokemon['Class 1'] == 'Grass', 'Attack']
rock_attack = pokemon_1.loc[pokemon['Class 1'] == 'Rock', 'Attack']
```

```python
#Testing the normality of distribution
shapiro_grass = shapiro(grass_attack)
shapiro_rock = shapiro(rock_attack)
```

```python
print(f"Shapiro-Wilk Test for Grass: p-value = {shapiro_grass.pvalue:.4f}")
print(f"Shapiro-Wilk Test for Rock: p-value = {shapiro_rock.pvalue:.4f}")
```

```python
from scipy.stats import ttest_ind, shapiro, levene
```

```python
#Testing homogeneity of variances
levene_test = levene(grass_attack, rock_attack)
print(f"Levene's test for equality of variances: p-value ={levene_test.pvalue:.4f}")
```

```python
#Student's t-test
t_stat, p_value = ttest_ind(grass_attack,rock_attack, alternative='greater')
print(f"t-stat: {t_stat:.4f}, p-value: {p_value:.4f}")
```

```python
alpha = 0.05
if p_value < alpha:
    print("Professor Oak is right: Grass class Pokémon have higher Normal Attack than Rock class Pokémon.")
else:
    print("There is no sufficient evidence to say that Grass class Pokémon are stronger than Rock class Pokémon in normal attack.")
```

```python
#Task 2
#Professor Oak has been unable to sleep at night for a long time, because he is worried about the question of whether it is true that Pokémon in the Water class are, on average, faster than Pokémon in the Normal class.
#Check if he is right and convince him of your conclusion statistically.
#Note: if there are Pokémon that belong to both classes, throw them away;
#You can assume that the distribution of movement speed is normal for all Pokémon classes.
```

```python
pokemon = pokemon[~((pokemon['Class 1'] == 'Water') & (pokemon['Class 2'] == 'Normal'))]
pokemon = pokemon[~((pokemon['Class 1'] == 'Normal') & (pokemon['Class 2'] == 'Water'))]
```

```python
water_speed = pokemon.loc[pokemon['Class 1'] == 'Water', 'Speed']
normal_speed = pokemon.loc[pokemon['Class 1'] == 'Normal', 'Speed']
```

```python
shapiro_water = shapiro(water_speed)
shapiro_normal = shapiro(normal_speed)
print(f"Shapiro-Wilk Test for Water: p-value = {shapiro_water.pvalue:.4f}")
print(f"Shapiro-Wilk Test for Normal: p-value = {shapiro_normal.pvalue:.4f}")
```

```python
levene_test = levene(water_speed, normal_speed)
print(f"Levene's test for equality of variances: p-value = {levene_test.pvalue:.4f}")
```

```python
t_stat, p_value = ttest_ind(water_speed, normal_speed, alternative='greater')
print(f"t-stat: {t_stat:.4f}, p-value: {p_value:.4f}")

# Вывод результата
alpha = 0.05
if p_value < alpha:
    print("Professor Oak is right: Water-class Pokemon are on average faster than Normal-class Pokemon.")
else:
    print("There is no sufficient evidence to say that Water-class Pokemon are faster than Normal-class Pokemon.")
```

```python
#Task 3
#Professor Oak is quite a mad scientist. He has invented a serum that is supposed to boost a Pokémon's speed. 
#However, we doubt the effectiveness of his vaccine. The professor administered this serum to the following Pokémon: see the array treated_pokemon. 
#Check whether his serum actually works and convince everyone of your conclusion statistically.
#You may assume that the speed distribution is normal for all Pokémon classes.
```

```python
treathed_pokemon = ['Mega Beedrill', 'Mega Alakazam', 'Deoxys Normal Forme', 'Mega Lopunny']
```

```python
treated_data = pokemon[pokemon['Name'].isin(treathed_pokemon)]
untreated_data = pokemon[~pokemon['Name'].isin(treathed_pokemon)]

treated_speeds = treated_data['Speed']
untreated_speeds = untreated_data['Speed']

shapiro_treated = shapiro(treated_speeds)
shapiro_untreated = shapiro(untreated_speeds)

print(f"Shapiro-Wilk test for speed (with serum): p-value = {shapiro_treated.pvalue:.4f}")
print(f"Shapiro-Wilk test for speed (without serum): p-value = {shapiro_untreated.pvalue:.4f}")
```

```python
levene_test = levene(treated_speeds, untreated_speeds)
print(f"Levene's test for equality of variances: p-value = {levene_test.pvalue:.4f}")

equal_var = levene_test.pvalue > 0.05 
t_stat, p_value = ttest_ind(treated_speeds, untreated_speeds, equal_var=equal_var)

print(f"t-stat: {t_stat:.4f}, p-value: {p_value:.4f}")
```

```python
alpha = 0.05
if p_value < alpha:
    print("The speeds of Pokémon that took the serum were statistically significantly different from the others.")
else:
    print("There is no sufficient evidence to suggest that the serum affects Pokemon speed.")
```

```python
#Task 4
#Professor Oak has always been fascinated by stories about legendary Pokémon. 
#However, he isn't entirely convinced that they are better than other Pokémon. Oak has asked us to investigate this.
#Check whether the sum of the characteristics HP, Attack, and Defense is indeed higher for legendary Pokémon compared to other Pokémon.
#What about the product of these same parameters?
#Find the answers to these questions and make a statistically convincing case for your conclusions.
#You may assume that the distribution of the sums and products of these parameters follows a normal distribution for all Pokémon classes.
```

```python
legendary = pokemon[pokemon['Legendary'] == True]
non_legendary = pokemon[pokemon['Legendary'] == False]
```

```python
legendary['Sum'] = legendary['HP'] + legendary['Attack'] + legendary['Defense']
non_legendary['Sum'] = non_legendary['HP'] + non_legendary['Attack'] + non_legendary['Defense']
```

```python
legendary['Product'] = legendary['HP'] * legendary['Attack'] * legendary['Defense']
non_legendary['Product'] = non_legendary['HP'] * non_legendary['Attack'] * non_legendary['Defense']
```

```python
shapiro_sum_legendary = shapiro(legendary['Sum'])
shapiro_sum_non_legendary = shapiro(non_legendary['Sum'])
shapiro_product_legendary = shapiro(legendary['Product'])
shapiro_product_non_legendary = shapiro(non_legendary['Product'])

print(f"Shapiro-Wilk test for sum (legendary): p-value = {shapiro_sum_legendary.pvalue:.4f}")
print(f"Shapiro-Wilk test for sum (non legendary): p-value = {shapiro_sum_non_legendary.pvalue:.4f}")
print(f"Shapiro-Wilk test for product (legendary): p-value = {shapiro_product_legendary.pvalue:.4f}")
print(f"Shapiro-Wilk test for product (non legendary): p-value = {shapiro_product_non_legendary.pvalue:.4f}")
```

```python
levene_sum = levene(legendary['Sum'], non_legendary['Sum'])
levene_product = levene(legendary['Product'], non_legendary['Product'])

print(f"Levene's test for sum: p-value = {levene_sum.pvalue:.4f}")
print(f"Levene's test for product: p-value = {levene_product.pvalue:.4f}")
```

```python
from scipy.stats import mannwhitneyu

u_stat_sum, p_value_sum = mannwhitneyu(legendary['Sum'], non_legendary['Sum'], alternative='greater')

u_stat_product, p_value_product = mannwhitneyu(legendary['Product'], non_legendary['Product'], alternative='greater')

print(f"Mann-Whitney test for sum: U-stat = {u_stat_sum:.4f}, p-value = {p_value_sum:.4f}")
print(f"Mann-Whitney test for product: U-stat = {u_stat_product:.4f}, p-value = {p_value_product:.4f}")
```

```python
alpha = 0.05
if p_value_sum < alpha:
    print("The sum of characteristics for legendary Pokémon is statistically significantly higher than for non-legendary Pokémon.")
else:
    print("There is not enough evidence to conclude that the sum of characteristics for legendary Pokémon is higher than for non-legendary Pokémon.")

if p_value_product < alpha:
    print("The product of characteristics for legendary Pokémon is statistically significantly higher than for non-legendary Pokémon.")
else:
    print("There is not enough evidence to conclude that the product of characteristics for legendary Pokémon is higher than for non-legendary Pokémon.")
```
